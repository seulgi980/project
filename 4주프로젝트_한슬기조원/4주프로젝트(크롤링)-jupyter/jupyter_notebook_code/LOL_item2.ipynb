```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from tqdm import tqdm


driver = webdriver.Chrome('./chromedriver.exe')
driver

html_lst = []


for user_name in tqdm(name_lst1100[:1]):
    # 페이지이동
    try:
        driver.get(f'https://poro.gg/summoner/kr/{user_name}%20')
        sleep(1)
     
        # 더보기 버튼 클릭   

        more_bt = driver.find_element(
                         By.CSS_SELECTOR,
                        'div.page-container > div:nth-child(2) > div:nth-child(6) > button')

        more_bt.click()     
        
        sleep(3)

        bt = driver.find_elements(
            By.CSS_SELECTOR,
            'div.page-container > div:nth-child(2) > div:nth-child(6) > div.mb-8px > div:nth-child(1) > div:nth-child(1) >  button' )  

        for i in range(len(bt)):
            try:
                bt[i].click()
            except:
                continue
                
        
        sleep(1)

        
        bt2 = driver.find_elements(
            By.CSS_SELECTOR,
            'div.page-container > div:nth-child(2) > div:nth-child(6) > div.mb-8px > div:nth-child(3) > div:nth-child(1) > ul.flex > li:nth-child(2) > button' )  

        for i in range(len(bt)):
            try:
                bt2[i].click()
                bt2[i].click()
               
            except:
                continue
                
        html = driver.page_source
        html_lst.append(html)
            
    except:
        print(f"{user_name} 닉변")
        continue
```

```python
from bs4 import BeautifulSoup
from tqdm import tqdm

cham_lst = []
item_lst = []
lst_10 = []
# 된거라도 먼저하자..

for i in tqdm(range(len(html_lst))):
    soup = BeautifulSoup(html_lst[i])

    # 내가픽한챔프
    ch_lst = []
    for a in range(len(soup.select('.champion > .icon '))): # 40게임
        s = soup.select('.champion > .icon ')[a].select('img')[0]['alt']
        ch_lst.append(s)
        
    cham_lst.append(ch_lst)

        
    # 아이템빌드
    items = [] # [ [1판아이템모음] [2판아이템모음] ,,,[40판아이템모음]]
    for b in range(len(soup.select('.bg-white > div:nth-child(1) > div:nth-child(1) > div:nth-child(1) > div:nth-child(1) > .relative >.has-tooltip '))): #40게임
        ith_game = soup.select('.css-xda5ep')[b]
        item_names = ith_game.select('img')
        lst = [] # [1판아이템1 1판아이템2~ 23개정도 1판아이템모음]
        for k in range(len(item_names)):
            item_name2 = item_names[k]['alt']
            lst.append(item_name2)
        
        items.append(lst)
    item_lst.append(items)
        
    # 10명챔프   
    lst_10s = []
    for d in range(len(soup.select('.participants'))):   
        lst_p = []
        for e in range(10):
            p = soup.select('.participants')[d].select('img')[e]['alt']
            lst_p.append(p)
        
        lst_10s.append(lst_p)
        
        
    lst_10.append(lst_10s)
        
```
