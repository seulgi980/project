# 인코더

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding, Conv1D, GlobalAveragePooling1D, Concatenate, Dense
class Encoder(tf.keras.Model):

  def __init__(self, vocab_size=2000): # 챔피언 이름개수 162개
    super(Encoder, self).__init__()
    self.emb = Embedding(vocab_size, 128)
    # 텍스트(정수인코딩된거) > 벡터화 해주기 = " --10개--" 를 embedding테이블을 이용해 lookuptable화 시키기
    # lookuptable은 (10,128)

    self.conv1 = Conv1D(filters = 64, kernel_size = 1, padding='valid', activation='tanh', strides=1) # 내챔피언
    self.conv2 = Conv1D(filters = 64, kernel_size = 4, padding='valid', activation='tanh', strides=1)  # 4명우리팀
    self.conv3 = Conv1D(filters = 64, kernel_size = 5, padding='valid', activation='tanh', strides=1) # 5명 상대팀
    self.pool = GlobalAveragePooling1D()
    self.concat = Concatenate()
    self.dense = Dense(512, activation = 'relu')

  def call(self, x): # seqs (32,10)가 들어와서
    net = self.emb(x) # seqs가 임베딩되면 (32, 10, 128)
    c1 = self.conv1(net[:,0:1,:])
    c2 = self.conv2(net[:,1:5,:])
    c3 = self.conv3(net[:,5:,:])
    p1 = self.pool(c1)
    p2 = self.pool(c2)
    p3 = self.pool(c3)
    concat_result = self.concat([p1,p2,p3])
    h = self.dense(concat_result)
    # Dense 하나 넣어주기 encoder contextvector와 decoder의 h0개수를 맞추기 위해

    return h
```

# 디코더 GRU사용
## 훈련할땐 티쳐포싱

```python
from tensorflow.keras.layers import GRU

class Decoder(tf.keras.Model):
  
  def __init__(self, vocab_size=2000): 
    super(Decoder, self).__init__()

    self.emb = Embedding(vocab_size, 128)
    self.gru = GRU(512, return_sequences=True, return_state=True)
    self.dense = Dense(vocab_size, activation='softmax')
    # 디코더 층 하나하나마다 soft맥스 사용(아래 민트색부분에서 사용할거)

  def call(self, inputs):
    h, y = inputs  # h : context vector , y : 타깃(아이템)
    # GRU라서 cell state가 없음⭐
    
    y = self.emb(y)

    y, h = self.gru(y, initial_state=h) #  initial_state=h 이전스텝의 h 강제로 넣어주기

    return self.dense(y), h # soft max결과와, h반환
           # genarater부분
```

# seq2seq구현

```python
class Seq2Seq(tf.keras.Model):

  def __init__(self, sos="<sos>", eos="<eos>"):
    super(Seq2Seq, self).__init__()
    self.enc = Encoder()
    self.dec = Decoder()
    self.sos = sos
    self.eos = eos

  def call(self, inputs, training=False):
    # 훈련땐 교사강요할거고
    if training == True:
      x, items = inputs # x = "챔피언10명", items = 아이템빌드(타겟)
      h = self.enc(x)
      # 인코더 돌려서 h(context vector얻어내고)
      # items(실제정답)과 함께 디코더에 투입 (교사강요)
      y, _ = self.dec((h, items))
      # 교사강요하기때문에 각 스텝의 입력값을 실제정답 items로 입력

      return y # soft맥스한 결과값



    # test땐 교사강요 안할거기 때문에!
    else:
      x = inputs
      h = self.enc(x)

      # <sos> 텐서 강제로 만들어주기
      # 처음에 y로 넣을게 없어서
      y = tf.convert_to_tensor(self.sos)
      y = tf.reshape(y, (1,1)) # y = [['시작']]

      # 예측값을 저장할 텐서 미리 만들어놓기
      total_seq = tf.TensorArray(tf.int32, 32) # 최대 32개의 공간을 미리 만들어 둔다.
      # 이게 아이템 32개까지 담을공간인가..?

      for idx in tf.range(32):
        y, h = self.dec([h,y]) # y는 이전스텝의 예측값
        y = tf.cast(tf.argmax(y, axis=-1), dtype=tf.int32) # softmax값 젤 큰거 인덱스추출

        y = tf.reshape(y, (1,1))
        total_seq = total_seq.write(idx, y) # 인덱스에 y값 채우기

        # y의 출력값이 eos라면 반복 중지
        if y == self.eos:
          break

      return tf.reshape(total_seq.stack(), (1, 32))
```

```python
# 훈련과 테스트 루틴 정의
@tf.function
def train_step(model, inputs, labels, loss_object, optimizer, train_loss, train_accuracy):
    output_labels = labels[:, 1:] # 제일 처음은 빼고 집어 넣는다. <sos>는 빼고~. 즉 <sos>가 없고 <eos>가 있고
    shifted_labels = labels[:, :-1] # <sos>를 포함하고 <eos>를 뺀다
    with tf.GradientTape() as tape:

    # 순전파
      predictions = model([inputs, shifted_labels], training=True)
      loss = loss_object(output_labels, predictions)

    # 역전파
    gradients = tape.gradient(loss, model.trainable_variables)

    # 최적화
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

    # 훈련결과 기록해놓기
    train_loss(loss)
    train_accuracy(output_labels, predictions)

@tf.function
def test_step(model, inputs):
    return model(inputs, training=False) # 테스트는 교사강요 X
```

```python
# word_index는 현재 모델 내부에 계산된 단어 인덱스를 "dictionary 형태로" 확인한다.
# 시작(4), 끝(5) 잘들어감
tokenizer_item.word_index
```

```python
# 모델 생성
model = Seq2Seq(sos=tokenizer_item.word_index['시작'],
                eos=tokenizer_item.word_index['끝'])
# word_index는 현재 모델 내부에 계산된 단어 인덱스를 "dictionary 형태로" 확인한다.

# Loss, 최적화 정의
loss_object = tf.keras.losses.SparseCategoricalCrossentropy()
optimizer = tf.keras.optimizers.Adam()

# 훈련 평가방식 정의
train_loss = tf.keras.metrics.Mean(name='train_loss')
train_accuracy = tf.keras.metrics.SparseCategoricalAccuracy(name='train_accuracy')
```

```python
train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).shuffle(10000).batch(64).prefetch(1024)
test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test)).batch(1).prefetch(1024)

# prefectch를 사용하면 학습 데이터를 나눠서 읽어오기 때문에 첫 번째 데이터를 GPU에서 학습하는 동안
# 두 번째 데이터를 CPU에서 준비할 수 있어 리소스의 유휴 상태를 줄일 수 있다.
# prefetch를 통해 미리 데이터를 올려놓기때문에 bottleneck현상을 줄일 수 있다.
```

```python
# 1분30초정도소요
from tqdm import tqdm
for epoch in tqdm(range(10)):
  for seqs, labels in train_ds:
    # print 해보니 seqs(64,10), labels은 (64,61)형태
    train_step(model, seqs, labels, loss_object, optimizer, train_loss, train_accuracy)
  print('Epoch {}, Loss: {}, Accuracy: {}'.format(epoch + 1,train_loss.result(),train_accuracy.result() * 100))

  train_loss.reset_states()
  train_accuracy.reset_states()
```
