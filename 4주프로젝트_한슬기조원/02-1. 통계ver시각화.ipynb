```python
# 메인룬
from tqdm import tqdm
for cham in tqdm([champion_name_eng[35]]):
  globals()[f"{cham}_mainlun"] = []
  TFdf = total_df['champion'] == cham
  for i in range(290060):
    if TFdf.iloc[i] == True:
      globals()[f"{cham}_mainlun"].append(total_df['mainlun'].iloc[i])
```

```python
# 보조룬
for cham in tqdm([champion_name_eng[35]]):
  globals()[f"{cham}_sublun"] = []
  TFdf = total_df['champion'] == cham
  for i in range(290060):
    if TFdf.iloc[i] == True:
      globals()[f"{cham}_sublun"].append(total_df['sublun'].iloc[i])
```

```python
# 스펠
for cham in tqdm([champion_name_eng[35]]):
  globals()[f"{cham}_spell"] = []
  TFdf = total_df['champion'] == cham
  for i in range(290060):
    if TFdf.iloc[i] == True:
      globals()[f"{cham}_spell"].append(total_df['spell'].iloc[i])
```
