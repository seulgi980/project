## Rating을 구분해주었습니다.
- 전체관람가 : PG /  ALL /  G / TV-PG / TV-Y / TV-G / ALL_AGES
- 7세관람가 : 7+ / TV-Y7 / TV-Y7-FV
- 13세관람가 : PG-13 / 13+
- 14세관람가 : TV-14
- 16세관람가 : 16+ / AGES_16_ / 16
- 17세 관람가 : R / TV-MA / NC-17'
- 18세관람가 : AGES_18_ / 18+
- 등급보류 아직 받지않음 : NR / UNRATED / NOT_RATE / TV-NR / NOT RATED/ UR

```python
rating_df['rating'].replace(["PG", "G", "TV-PG", "TV-Y", "TV-G", "ALL_AGES"], ['ALL','ALL','ALL','ALL','ALL','ALL'])
```

```python
rating_df['rating'].replace(['TV-Y7', 'TV-Y7_FV'], ['7+','7+'])
```

```python
# 전체관람가 / 7세이상 / 13세이상 / 14세이상 / 16세이상 / 17세이상 / 18세이상 / unrated

rating_df['rating'] = rating_df['rating'].replace(["PG", "G", "TV-PG", "TV-Y", "TV-G", "ALL_AGES"], ['ALL','ALL','ALL','ALL','ALL','ALL'])
rating_df['rating'] = rating_df['rating'].replace(['TV-Y7', 'TV-Y7-FV', '7+'], ['07+','07+', '07+'])
rating_df['rating'] = rating_df['rating'].replace('PG-13', '13+')
rating_df['rating'] = rating_df['rating'].replace('TV-14', '14+')
rating_df['rating'] = rating_df['rating'].replace(['AGES_16_', '16'], ['16+', '16+'])
rating_df['rating'] = rating_df['rating'].replace(['R', 'TV-MA', 'NC-17'], ['17+', '17+', '17+'])
rating_df['rating'] = rating_df['rating'].replace('AGES_18_', '18+')
rating_df['rating'] = rating_df['rating'].replace(['NR', 'NOT_RATE', 'TV-NR', 'NOT RATED', 'UR'], ['UNRATED', 'UNRATED', 'UNRATED', 'UNRATED','UNRATED'])

rating_df = rating_df.sort_values(by='rating')
rating_df
```
