# 가사 한번에 가져오기!
- 하나하나 가져오기 불편하여서 함수를 만드는법을 생각해보았습니다.

## 함수로 노래가사 가져오기

```python
X = []
y = []

def crol(url, label):
  # 1. 크롤링
  headers = {'user-agent' : UserAgent().chrome}
  response = requests.get(url, headers=headers)

  page = response.content
  soup = BeautifulSoup(page, 'html.parser')

  lyric = soup.select_one('.lyricsContainer xmp').text

  # 2. 전처리
  lyric = lyric.split('\r\n')
  lyric = ''.join(lyric)

  okt = Okt()
  lyric = okt.morphs(lyric, norm=True, stem=True)

  # 3. 다시조립
  lyric = ' '.join(lyric)

  X.append(lyric)
  y.append(label)
```

```python
# 이런식으로 하나하나 넣어주는 방법
# 위의 방법은 원하는 데이터를 골라넣거나, 하나하나 추가할때 유용할 것 같습니다.

crol('https://music.bugs.co.kr/track/6179174?wl_ref=list_tr_08_search', 0)
```

## 편리성을 위해 페이지에 있는 차트를 통째로 가져오는 방법을 선택하였습니다.

image.png

```python
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
import requests
from konlpy.tag import Okt

X = []
y = []


def crol(url, genre):
  URL_lst = []
  headers = {'user-agent' : UserAgent().chrome}
  response = requests.get(url, headers=headers)
  page = response.content
  soup = BeautifulSoup(page, 'html.parser')

  for i in range(100):
    URL = soup.select("a.trackInfo")[i]['href']
    URL_lst.append(URL)

  for url in URL_lst:
    headers = {'user-agent' : UserAgent().chrome}
    response = requests.get(url, headers=headers)
    page = response.content
    soup = BeautifulSoup(page, 'html.parser')

    lyric = soup.select_one('.lyricsContainer xmp').text

    # 2. 전처리
    lyric = lyric.split('\r\n')
    lyric = ''.join(lyric)

    okt = Okt()
    lyric = okt.morphs(lyric, norm=True, stem=True)

    # 3. 다시조립
    lyric = ' '.join(lyric)

    X.append(lyric)

  if genre == '발라드':
    for g in range(100):
      y.append(1)

  elif genre == '댄스':
    for g in range(100):
      y.append(0)
```

```python
crol('https://music.bugs.co.kr/genre/chart/kpop/ballad/total/day', '발라드')
```

```python
crol('https://music.bugs.co.kr/genre/chart/kpop/dance/total/day', '댄스')
```

# X, y를 train, valid, test셋으로 분리
- 검증성을 높이기위해 검증(valid)셋을 사용하였습니다.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2,
    stratify = y,
    random_state=42
)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(
    X_train,
    y_train,
    test_size = 0.2,
    stratify = y_train,
    random_state=42
)
```

```python
len(X_train), len(X_valid), len(X_test)
```

# 4. CountVectorizer, TfidfVectorizer를 이용해 벡터화
-  TfidfVectorizer 을 선택하였습니다.

```python
# 모델에 numpy배열로 넣어주기 위해 y를 넘파이변환
import numpy as np

y_train = np.array(y_train)
y_valid = np.array(y_valid)
y_test = np.array(y_test)
```

```python
y_train.shape, y_valid.shape, y_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect= TfidfVectorizer()

X_train_tfidf_vect = tfidf_vect.fit_transform(X_train)

 # 검증세트는 꼭 transform만해주기⭐
X_valid_tfidf_vect = tfidf_vect.transform(X_valid)
X_test_tfidf_vect = tfidf_vect.transform(X_test)

X_train_tfidf_vect
```

```python
X_train_tfidf_vect.shape, X_valid_tfidf_vect.shape, X_test_tfidf_vect.shape
```

# 5. lightGBM으로 모델만들기

```python
!pip install lightgbm==3.3.2
```

```python
import lightgbm
from lightgbm import LGBMClassifier

lgbm_wrapper = LGBMClassifier(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth = 5
    )
```


# 훈련

```python
evals = [
    (X_train_tfidf_vect, y_train),
    (X_valid_tfidf_vect, y_valid)
]


lgbm_wrapper.fit(
    X_train_tfidf_vect,
    y_train,
    early_stopping_rounds=500,
    eval_metric='logloss',
    eval_set=evals
)
```

# test셋으로 평가

```python
from sklearn.metrics import accuracy_score

pred = lgbm_wrapper.predict(X_test_tfidf_vect)

accuracy_score(y_test , pred)
```

```python
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
import requests
from konlpy.tag import Okt

def crol(url, genre):
  URL_lst = []
  headers = {'user-agent' : UserAgent().chrome}
  response = requests.get(url, headers=headers)
  page = response.content
  soup = BeautifulSoup(page, 'html.parser')

  for i in range(23):
    URL = soup.select("a.trackInfo")[i]['href']
    URL_lst.append(URL)

  for url in URL_lst:
    headers = {'user-agent' : UserAgent().chrome}
    response = requests.get(url, headers=headers)
    page = response.content
    soup = BeautifulSoup(page, 'html.parser')

    lyric = soup.select_one('.lyricsContainer xmp').text

    # 2. 전처리
    lyric = lyric.split('\r\n')
    lyric = ''.join(lyric)

    okt = Okt()
    lyric = okt.morphs(lyric, norm=True, stem=True)

    # 3. 다시조립
    lyric = ' '.join(lyric)

    X.append(lyric)

  if genre == '발라드':
    for g in range(23):
      y.append(1)

  elif genre == '댄스':
    for g in range(23):
      y.append(0)
```
